Converters

```python
def rename(name):
    if name=="Setosa":
        return 'S'
    elif name=="Virginica":
        return 'V'
    else:
        return name
```

```python
rename('Setosa')
```

```python
pd.read_csv('dataset\\iris.tsv',sep='\t',converters={'Setosa':rename,'Virginica':rename})
```

loading a huge dataset in chunks
